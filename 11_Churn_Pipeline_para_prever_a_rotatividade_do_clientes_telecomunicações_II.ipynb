# Introdução

Você pode prever os clientes que provavelmente deixarão a empresa?

Ao escolher um provedor de serviços de telefonia, os clientes geralmente têm muitas opções. Eles podem escolher qualquer provedor de serviços e podem se afastar do provedor atual. 

Quando um cliente opta por mudar do provedor atual para um novo provedor, isso resulta em perda de negócios e receita para o provedor atual. **O percentual de clientes que saem e desconectam do serviço é conhecido como “churn”.** 


**Uma base de clientes estável é a chave para o sucesso de qualquer negócio.**

As empresas tentam manter os clientes satisfeitos, para retê-los o maior tempo possível. No entanto, no mundo real, a rotatividade de clientes pode chegar a 25% ao ano no setor de telecomunicações. Além disso, o custo de adquirir um novo cliente é 10 vezes maior do que o custo de reter um cliente existente. Isso representa um sério desafio para os proprietários de empresas.



## Contexto: CHURN

A rotatividade de clientes é uma das métricas mais importantes para uma empresa em crescimento avaliar. Embora não seja a medida mais animadora, é um número que pode dar à sua empresa a dura verdade sobre a retenção de clientes.

É difícil medir o sucesso se você não medir as falhas inevitáveis ​​também. Embora você se esforce para que 100% dos clientes fiquem com sua empresa, isso é simplesmente irreal. É aí que entra a rotatividade de clientes.

A rotatividade de clientes é a porcentagem de clientes que pararam de usar o produto ou serviço de sua empresa durante um determinado período. Você pode calcular a taxa de rotatividade dividindo o número de clientes perdidos durante aquele período - digamos um trimestre - pelo número de clientes que você tinha no início do período.

Por exemplo, se você iniciar o trimestre com 400 clientes e terminar com 380, sua taxa de desligamento será de 5% porque você perdeu 5% dos clientes.

Obviamente, sua empresa deve almejar uma taxa de churn o mais próximo possível de 0%. Para fazer isso, sua empresa deve estar sempre no topo da taxa de rotatividade e tratá-la como uma prioridade.

### Por que a taxa de rotatividade do cliente é importante?

Você pode estar se perguntando por que é necessário calcular a taxa de rotatividade. Naturalmente, você vai perder alguns clientes aqui e ali, e 5% não parece tão ruim, certo?

Bem, é importante porque custa mais adquirir novos clientes do que reter os clientes existentes. Na verdade, um aumento de apenas 5% na retenção de clientes pode gerar um aumento de pelo menos 25% no lucro. Isso ocorre porque os clientes recorrentes provavelmente gastarão 67% a mais nos produtos e serviços da sua empresa. Como resultado, sua empresa pode gastar menos com os custos operacionais de ter que adquirir novos clientes. Você não precisa gastar tempo e dinheiro para convencer um cliente existente a escolher sua empresa em vez dos concorrentes, porque eles já tomaram essa decisão.

Novamente, pode parecer que uma taxa de rotatividade de 5% é sólida e saudável. Você ainda pode ter uma grande receita com essa taxa de rotatividade.

### Por que os clientes mudam?

* **O cliente não quer mais seu produto.**

O que quer que tenha atraído o cliente em primeiro lugar, acabou. Quer sejam os valores da empresa, a qualidade do produto ou o preço em que o produto foi apresentado pela primeira vez.

* **Experiência do usuário ruim**

Isso pode ser uma atualização de IU incorreta, falta de recursos importantes ou até mesmo algo tão simples como um pequeno bug.

* **O cliente encontrou uma alternativa melhor**

Este é provavelmente o mais comum, pois os clientes estão sempre em busca de um negócio melhor. 

### Como reduzir a rotatividade do cliente

* Concentre sua atenção em seus melhores clientes.
* Analise a rotatividade conforme ela ocorre.
* Mostre a seus clientes que você se importa.
* Invista em seus melhores clientes
* Peça feedback

## 2. Pré-processamento - Vamos limpar nossos dados

### Criação de um pipeline

Nosso pipeline consistirá nas etapas de pré-processamento seguidas por um estimador no final do pipeline. Esta é uma convenção geral, já que os pipelines do scikit learn chamarão fit_transform todas as etapas do pipeline, exceto a última, onde apenas chamará fit. Dependendo de nossos requisitos, podemos incluir, por exemplo, imputação, dimensionamento, codificação e redução de dimensionalidade.

#### **Por que usar um pipeline para pré-processamento?**


##### **Ele evita o vazamento de dados** 

O vazamento de dados acontece quando incluímos informações dos dados que estamos tentando prever durante o treinamento. Isso acontece, por exemplo, quando calculamos a média de imputação para todo o conjunto de dados, em vez de apenas para os dados de treinamento. Nossa estimativa de desempenho seria então excessivamente otimista. Um pipeline torna muito fácil evitar o vazamento de dados, pois chamamos *fit* os dados de treinamento e predictos dados de teste.

##### **Minimiza erros**

Com o pré-processamento manual, há muito mais etapas menores envolvidas e, às vezes, cada variável precisa ser feita individualmente. Um pipeline simplifica esse processo e minimiza a possibilidade de erros.

##### **Código de aparência mais limpo**

O pipeline é escrito uma vez e, em seguida, só precisamos chamar o método correspondente. Como resultado, o código é mais limpo e fácil de ler.

##### Validação cruzada em todo o pipeline**

Em vez de validação cruzada apenas do estimador, podemos fazer a validação cruzada de todo o pipeline. O mesmo se aplica ao ajuste de hiperparâmetros.


#### **Etapas de pré-processamento em nosso pipeline**

##### **Imputação**

Descobrimos algumas cadeias de caracteres vazias *TotalCharges* e as passamos para NaN. Nós usaremos o aprendizado do sci-kit SimpleImputer.

##### *Scaling*

Os algoritmos de escalonamento que medem distâncias entre pontos de dados funcionam muito melhor quando os dados são escalados. A ideia é fazer diferentes intervalos numéricos comparáveis ​​uns aos outros. Usaremos o scikit-learn StandardScaler.

##### **Codificação categórica**

Vimos anteriormente que nosso conjunto de dados contém principalmente variáveis ​​categóricas. Como a maioria dos algoritmos de aprendizado de máquina não entende dados não numéricos, precisaremos convertê-los em numéricos. Como todos os recursos têm uma cardinalidade muito baixa, escolheremos o OneHotEncoder.

##### **Seleção de recursos**

Nosso pipeline também incluirá uma parte de seleção de recursos, onde selecionaremos apenas os recursos de melhor desempenho. Usaremos o scikit learn's SelectKBest.

##### **Construindo o pipeline**

Pense no *preprocessor* como uma coleção de pipelines, cada um projetado para lidar com um tipo específico de dados - um para recursos numéricos e outro para recursos categóricos. *ColumnTransformer* nos permite aplicar seletivamente as etapas de pré-processamento.

```python
# Criação do pipeline no scikit para pré-processamento

## Selecionando as features categóricos e numéricos
numerical_ix = X.select_dtypes(include=np.number).columns
categorical_ix = X.select_dtypes(exclude=np.number).columns

## Crie pipelines de pré-processamento para cada tipo de dados 
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('encoder', OrdinalEncoder()),
    ('scaler', StandardScaler())])

## Juntando as etapas de pré-processamento
preprocessor = ColumnTransformer([
        ('numerical', numerical_transformer, numerical_ix),
        ('categorical', categorical_transformer, categorical_ix)],
         remainder='passthrough')
```

A seguir, vamos experimentar diferentes algoritmos de classificação criando uma lista e fazendo um loop por essa lista. O pipeline final, portanto, será criado dentro do loop for.

#### **Criação de um exemplo de pipeline com o classificador kNN**

Para fins de demonstração, criaremos um pipeline "totalmente funcional" como você faria normalmente, ou seja, com um estimador no final.

**Chamando .fit () no pipeline.**

Observe que apenas chamamos .fit () no pipeline, em vez de fit_transform (). Isso ocorre porque, como usuário, chamamos apenas .fit (), mas internamente, ele também chamará .transform () - exceto na última etapa. É por isso que é tão importante, como mencionado antes, incluir o estimador no final do pipeline. Observe também que o pipeline só funcionará com objetos que suportam .fit e .transform ().

```python
# Criando um exemplo de pipeline com kNN

example_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('selector', SelectKBest(k=len(X.columns))), 
    ('classifier', KNeighborsClassifier())
])

cross_val_score(example_pipe, X, y, cv=5, scoring='roc_auc').mean()
#example_pipe.fit(X_train, y_train)
#example_pipe.score(X_test, y_test)
```

```python
# Vamos tentar o kNN, mas modificando a quantidade de recursos que incluímos e comparar os resultados

k_range = list(range(1, len(X.columns))) # 1 até 18
k_scores = []

for k in k_range:
  example_pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('selector', SelectKBest(k=k)), 
    ('classifier', KNeighborsClassifier())])
  score = cross_val_score(example_pipe, X, y, cv=5, scoring='roc_auc').mean()
  k_scores.append(score)

k_scores_df = pd.DataFrame({
    'k_range': 
      k_range, 
    'k_scores':
      k_scores
})

max_value_k = np.array(k_scores_df[k_scores_df['k_scores'] == k_scores_df['k_scores'].max()])

print('Obtemos a maior AUC validada cruzada com {} selecionando k={}.\n\n'.format(round(max_value_k[0,1],4), max_value_k[0,0].astype(int)))

# Plotting the AUC across the different k values
plt.plot(k_range, k_scores)
plt.xlabel('Value of k in SelectKBEST')
plt.xticks(k_range, rotation=20) 
plt.ylabel('Cross-Validated AUC')
```

### Encontrando o melhor modelo 


A ideia é encontrar o algoritmo com melhor desempenho em seu nível de linha de base, ou seja, sem ajustes, e selecioná-lo para melhorá-lo ainda mais por meio do ajuste de hiperparâmetros. Vamos percorrer diferentes algoritmos e ver qual tem o melhor desempenho - Observe que estamos usando roc_auc como uma métrica de avaliação!

```python
# Experimentando diferentes classificadores e selecionando os melhores

## Crie uma lista de classificadores que iremos percorrer
classifiers = [
    KNeighborsClassifier(),
    SVC(random_state=123),
    DecisionTreeClassifier(random_state=123),
    RandomForestClassifier(random_state=123),
    AdaBoostClassifier(random_state=123),
    GradientBoostingClassifier(random_state=123)
    ]

classifier_names = [
    'KNeighborsClassifier()',
    'SVC()',
    'DecisionTreeClassifier()',
    'RandomForestClassifier()',
    'AdaBoostClassifier()',
    'GradientBoostingClassifier()'
]

model_scores = []

## Looping through the classifiers
for classifier, name in zip(classifiers, classifier_names):
    pipe = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('selector', SelectKBest(k=len(X.columns))),
        ('classifier', classifier)])
    score = cross_val_score(pipe, X, y, cv=10, scoring='roc_auc').mean() # we're using the AUC here!
    model_scores.append(score)
    print("Model score for {}: {}".format(name, score))
```

Agora é hora de comparar as pontuações:

```python

# Plotting models and performance

plt.plot(classifier_names, model_scores)
plt.xlabel('Classifier Algorithms')
plt.xticks(classifier_names, rotation=45) 
plt.ylabel('Cross-Validated AUC')
```

```python
# Criação de df com desempenho de modelo

model_performance = pd.DataFrame({
    'Classifier': 
      classifier_names, 
    'Cross-validated AUC':
      model_scores
})

## Classificando de maneira decrescente
sorted_models = model_performance.sort_values('Cross-validated AUC', ascending = False, ignore_index=True)
display(sorted_models)
## Extrair o modelo de melhor desempenho e salvá-lo para a etapa seguinte
best_model = eval(sorted_models['Classifier'][0]) # usamos eval (), porque caso contrário, obteríamos uma string em vez da classe de algoritmo
```

## Ajuste de hiperparâmetros

Agora que identificamos nosso melhor candidato, podemos prosseguir para o ajuste de hiperparâmetros. Essencialmente, isso significa encontrar as melhores “configurações” do algoritmo.

**Como sabemos o que ajustar?**

Podemos alterar muitos hiperparâmetros em todo o pipeline. Vamos experimentar todos eles? Não! A validação cruzada é "cara" e, portanto, nos concentraremos em um número limitado de hiperparâmetros. Podemos obter uma lista completa de todos os hiperparâmetros com *pipe.get_params().keys()*.


A ideia por trás *GridSearchCV* é que especificamos os parâmetros e seus valores correspondentes que queremos pesquisar. O GridSearch tentará todas as combinações possíveis e determinará qual teve o melhor desempenho.

Em suma: escolheremos o melhor modelo de nosso dataframe e realizaremos uma pesquisa com o GridSearchCV do scikit learn para encontrar as melhores "configurações" (hiperparâmetros) para os algoritmos. Iremos também, simultaneamente, procurar o melhor k em nossa etapa de seleção de recursos com SelectKBest.

```python
## criando o pipeline final 
pipe = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('selector', SelectKBest(k=len(X.columns))),
        ('classifier', best_model)
])

score = cross_val_score(pipe, X, y, cv=10, scoring='roc_auc').mean()
print('O pipeline com a {}, melhor baseline, tem um cross-validated AUC de {}%'.format(str(best_model), score))
```

Vamos dar uma olhada em quais parâmetros podemos realmente ajustar. Isso será importante quando criarmos o modelo.

```python
pipe.get_params().keys()
```

**Quais são, então, alguns parâmetros que queremos ajustar em nosso pipeline?**

**Em SelectKBest:**
* k

**Em GradientBoostingClassifier:**

* n_estimators
* learning_rate
* max_depth

Lembre-se: quanto mais hiperparâmetros você pesquisar, mais tempo demorará.
Portanto, nos limitaremos aos quatro hiperparâmetros mencionados acima. Agora podemos simplesmente criar uma "grade" e realizar uma busca pela melhor combinação com GridSearchCV.

```python
# Criando o grid
grid = {
    "selector__k": [12,13,14,15,16,17,18],
    "classifier__max_depth":[1,3,5],
    "classifier__learning_rate":[0.01,0.1,1],
    "classifier__n_estimators":[100,200,300,400]
}

# Instanciando a pesquisa e ajustando os dados
gridsearch = GridSearchCV(estimator=pipe, param_grid=grid, n_jobs= 1, scoring='roc_auc') # GridSearchCV usa 5 CV por padrão
gridsearch.fit(X, y)  

# Obtendo os resultados
print(gridsearch.best_params_) 
print(gridsearch.best_score_)
```
