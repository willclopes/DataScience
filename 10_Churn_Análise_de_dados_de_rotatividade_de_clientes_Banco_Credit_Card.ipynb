# Introdução

Você pode prever os clientes que provavelmente deixarão a empresa?

Aumentar sua base de clientes é bom, mas não perdê-los é ainda melhor. As empresas estão cada vez mais preocupadas com a experiência do cliente, por isso gastam muito dinheiro em áreas que se dedicam exclusivamente a entender o comportamento dos clientes e a melhor forma de retê-los em sua base.

É evidente que não basta identificar quem tem probabilidade de sair da empresa, mas sim encontrar os principais motivadores que os levam a sair. Com essas informações em mãos, estratégias podem ser criadas para eliminar ou reduzir esses infratores.

Quando um cliente opta por mudar do serviço atual para um novo serviço de outra empresa, isso resulta em perda de negócios e receita para a empresa atual. **O percentual de clientes que saem e desconectam do serviço atual é conhecido como “churn”.** 


**Uma base de clientes estável é a chave para o sucesso de qualquer negócio.**

As empresas tentam manter os clientes satisfeitos, para retê-los o maior tempo possível. No entanto, no mundo real, a rotatividade de clientes pode chegar a um % muito alto por ano em deternimados setores. Além disso, o custo de adquirir um novo cliente é 10 vezes maior do que o custo de reter um cliente existente. Isso representa um sério desafio para os proprietários de empresas.



## Contexto: CHURN

A rotatividade de clientes é uma das métricas mais importantes para uma empresa em crescimento avaliar. Embora não seja a medida mais animadora, é um número que pode dar à sua empresa a dura verdade sobre a retenção de clientes.

É difícil medir o sucesso se você não medir as falhas inevitáveis ​​também. Embora você se esforce para que 100% dos clientes fiquem com sua empresa, isso é simplesmente irreal. É aí que entra a rotatividade de clientes.

A rotatividade de clientes é a porcentagem de clientes que pararam de usar o produto ou serviço de sua empresa durante um determinado período. Você pode calcular a taxa de rotatividade dividindo o número de clientes perdidos durante aquele período - digamos um trimestre - pelo número de clientes que você tinha no início do período.

Por exemplo, se você iniciar o trimestre com 400 clientes e terminar com 380, sua taxa de desligamento será de 5% porque você perdeu 5% dos clientes.

Obviamente, sua empresa deve almejar uma taxa de churn o mais próximo possível de 0%. Para fazer isso, sua empresa deve estar sempre no topo da taxa de rotatividade e tratá-la como uma prioridade.

### Por que a taxa de rotatividade do cliente é importante?

Você pode estar se perguntando por que é necessário calcular a taxa de rotatividade. Naturalmente, você vai perder alguns clientes aqui e ali, e 5% não parece tão ruim, certo?

Bem, é importante porque custa mais adquirir novos clientes do que reter os clientes existentes. Na verdade, um aumento de apenas 5% na retenção de clientes pode gerar um aumento de pelo menos 25% no lucro. Isso ocorre porque os clientes recorrentes provavelmente gastarão 67% a mais nos produtos e serviços da sua empresa. Como resultado, sua empresa pode gastar menos com os custos operacionais de ter que adquirir novos clientes. Você não precisa gastar tempo e dinheiro para convencer um cliente existente a escolher sua empresa em vez dos concorrentes, porque eles já tomaram essa decisão.

Novamente, pode parecer que uma taxa de rotatividade de 5% é sólida e saudável. Você ainda pode ter uma grande receita com essa taxa de rotatividade.

### Por que os clientes mudam?

* **O cliente não quer mais seu produto.**

O que quer que tenha atraído o cliente em primeiro lugar, acabou. Quer sejam os valores da empresa, a qualidade do produto ou o preço em que o produto foi apresentado pela primeira vez.

* **Experiência do usuário ruim**

Isso pode ser uma atualização de IU incorreta, falta de recursos importantes ou até mesmo algo tão simples como um pequeno bug.

* **O cliente encontrou uma alternativa melhor**

Este é provavelmente o mais comum, pois os clientes estão sempre em busca de um negócio melhor. 

### Como reduzir a rotatividade do cliente

* Analise a rotatividade conforme ela ocorre.

Agora, a intenção é preparar os dados. Começando com multicolinearidade. 

A Multicolinearidade é a ocorrência de altas intercorrelações entre duas ou mais variáveis ​​independentes em um modelo de regressão múltipla. A multicolinearidade pode levar a resultados distorcidos ou enganosos quando um pesquisador ou analista tenta determinar quão bem cada variável independente pode ser usada de forma mais eficaz para prever ou compreender a variável dependente em um modelo estatístico. [Fonte](https://www.investopedia.com/terms/m/multicollinearity.asp): 

A seguir, vamos criar uma função para identificar a multicolinearidade de variáveis ​​que existem em um determinado conjunto de dados e a aplicar ao conjunto de dados.




```python
# Função para detectar a multicolinearidade das variáveis ​​do conjunto de dados

def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

```python
# Aplicando a funcao ao dataset

analysisVIF = baseChurn.select_dtypes(['number']).drop(columns=['customerID'])

calc_vif(analysisVIF).sort_values(by='VIF', ascending=False).head(5)
```

Uma multicolinearidade muito alta é percebida para certas variáveis.

Podemos perceber isso quando plotamos o mapa do cluster pela primeira vez.

Pode-se perceber que xistem altas correlações entre as variáveis, no eixo diagonal que começa no canto superior esquerdo e termina no canto inferior direito. Vamos examinar isso mais de perto. Plotei as variáveis ​​com a maior multicolinearidade junto com as variáveis ​​com a maior correlação.

```python
# I already knew before the previous step that many variables would 
# have a high multicollinearity because in the cluster map it was very visible.

# The VIF function helped me to confirm my first suspicion.

# Now, as I know which variables I need to look closest, I'll do that.

groupColumns = [['Total_Revolving_Bal','Avg_Utilization_Ratio']
               ,['Total_Amt_Chng_Q4_Q1','Total_Ct_Chng_Q4_Q1']
               ,['Customer_Age','Months_on_book']
               ,['Credit_Limit','Avg_Open_To_Buy']
               ,['Total_Trans_Amt','Total_Trans_Ct']]

for i in range(len(groupColumns)):
  col1 = groupColumns[i][0]
  col2 = groupColumns[i][1]
  sns.relplot(x=analysisVIF[col1], y=analysisVIF[col2])

plt.show()
```

Para finalizar a implementação do modelo de propósito prático, vamos ao grande final : a implementação de técnicas de ajuste para reduzir o ajuste excessivo do modelo.

Ao treinar um modelo, ele deve encontrar um peso para cada recurso.

Cada peso é um valor no vetor teta.

Considerando que há uma penalidade em ter um peso para um recurso, isso incentiva o modelo a mover os pesos para mais perto de zero para alguns recursos.

Portanto, uma regularização minimiza a complexidade de um modelo para evitar overfitting.

```python
# 1a Tentativa:

penalty = ['l1', 'l2'] 

C = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

parameters = dict(C=C, penalty=penalty)
lr_classifier = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = "balanced_accuracy",
                           cv = 10,
                           n_jobs = -1)

lr_classifier  = lr_classifier .fit(X_train, y_train)
lr_best_accuracy = lr_classifier.best_score_
lr_best_parameters = lr_classifier.best_params_
lr_best_accuracy, lr_best_parameters
```

Ao fazer isso, obtivemos o seguinte retorno: (0,6575825285438619, {'C': 10, 'penalidade': '12'})

Em outras palavras, a penalidade mais eficiente foi 12.

Como percebemos que o l2 era a melhor penalidade, apliquei-a no modelo e enviei para a base de teste.

```python
penalty = ['l2']

C = [ 0.0001, 0.001, 0.01, 0.02, 0.05]

parameters = dict(C=C, penalty=penalty)
lr_classifier = GridSearchCV(estimator = classifier,
                           param_grid = parameters,
                           scoring = "balanced_accuracy",
                           cv = 10,
                           n_jobs = -1)
lr_classifier  = lr_classifier .fit(X_train, y_train)
lr_best_accuracy = lr_classifier.best_score_
lr_best_parameters = lr_classifier.best_params_
lr_best_accuracy, lr_best_parameters
```
