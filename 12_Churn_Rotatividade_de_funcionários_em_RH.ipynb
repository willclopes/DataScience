# Introdução

Você pode prever os clientes que provavelmente deixarão a empresa?

Ao escolher um provedor de serviços de telefonia, os clientes geralmente têm muitas opções. Eles podem escolher qualquer provedor de serviços e podem se afastar do provedor atual. 

Quando um cliente opta por mudar do provedor atual para um novo provedor, isso resulta em perda de negócios e receita para o provedor atual. **O percentual de clientes que saem e desconectam do serviço é conhecido como “churn”.** 


**Uma base de clientes estável é a chave para o sucesso de qualquer negócio.**

As empresas tentam manter os clientes satisfeitos, para retê-los o maior tempo possível. No entanto, no mundo real, a rotatividade de clientes pode chegar a 25% ao ano no setor de telecomunicações. Além disso, o custo de adquirir um novo cliente é 10 vezes maior do que o custo de reter um cliente existente. Isso representa um sério desafio para os proprietários de empresas.



## Contexto: CHURN

A rotatividade de clientes é uma das métricas mais importantes para uma empresa em crescimento avaliar. Embora não seja a medida mais animadora, é um número que pode dar à sua empresa a dura verdade sobre a retenção de clientes.

É difícil medir o sucesso se você não medir as falhas inevitáveis ​​também. Embora você se esforce para que 100% dos clientes fiquem com sua empresa, isso é simplesmente irreal. É aí que entra a rotatividade de clientes.

A rotatividade de clientes é a porcentagem de clientes que pararam de usar o produto ou serviço de sua empresa durante um determinado período. Você pode calcular a taxa de rotatividade dividindo o número de clientes perdidos durante aquele período - digamos um trimestre - pelo número de clientes que você tinha no início do período.

Por exemplo, se você iniciar o trimestre com 400 clientes e terminar com 380, sua taxa de desligamento será de 5% porque você perdeu 5% dos clientes.

Obviamente, sua empresa deve almejar uma taxa de churn o mais próximo possível de 0%. Para fazer isso, sua empresa deve estar sempre no topo da taxa de rotatividade e tratá-la como uma prioridade.

### Por que a taxa de rotatividade do cliente é importante?

Você pode estar se perguntando por que é necessário calcular a taxa de rotatividade. Naturalmente, você vai perder alguns clientes aqui e ali, e 5% não parece tão ruim, certo?

Bem, é importante porque custa mais adquirir novos clientes do que reter os clientes existentes. Na verdade, um aumento de apenas 5% na retenção de clientes pode gerar um aumento de pelo menos 25% no lucro. Isso ocorre porque os clientes recorrentes provavelmente gastarão 67% a mais nos produtos e serviços da sua empresa. Como resultado, sua empresa pode gastar menos com os custos operacionais de ter que adquirir novos clientes. Você não precisa gastar tempo e dinheiro para convencer um cliente existente a escolher sua empresa em vez dos concorrentes, porque eles já tomaram essa decisão.

Novamente, pode parecer que uma taxa de rotatividade de 5% é sólida e saudável. Você ainda pode ter uma grande receita com essa taxa de rotatividade.

### Por que os clientes mudam?

* **O cliente não quer mais seu produto.**

O que quer que tenha atraído o cliente em primeiro lugar, acabou. Quer sejam os valores da empresa, a qualidade do produto ou o preço em que o produto foi apresentado pela primeira vez.

* **Experiência do usuário ruim**

Isso pode ser uma atualização de IU incorreta, falta de recursos importantes ou até mesmo algo tão simples como um pequeno bug.

* **O cliente encontrou uma alternativa melhor**

Este é provavelmente o mais comum, pois os clientes estão sempre em busca de um negócio melhor. 

### Como reduzir a rotatividade do cliente

* Concentre sua atenção em seus melhores clientes.
* Analise a rotatividade conforme ela ocorre.
* Mostre a seus clientes que você se importa.
* Invista em seus melhores clientes
* Peça feedback
* Comunique-se frequentemente

## Pipeline de pré-processamento

Nesta seção, realizamos etapas de pré-processamento de dados para preparar os conjuntos de dados para a implementação do algoritmo de aprendizado de máquina. 

### Codificação

Os algoritmos de aprendizado de máquina geralmente podem ter apenas valores numéricos como suas variáveis ​​de previsão. Conseqüentemente, a codificação das features torna-se necessária, pois eles codificam as variáveis categóricas em valores numéricos. 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# Crie um objeto de codificador de rótulo
le = LabelEncoder()
```

```python
print(df_HR.shape)
df_HR.head()
```

```python
# A codificação de rótulo será usada para colunas com 2 ou menos valores únicos
le_count = 0
for col in df_HR.columns[1:]:
    if df_HR[col].dtype == 'object':
        if len(list(df_HR[col].unique())) <= 2:
            le.fit(df_HR[col])
            df_HR[col] = le.transform(df_HR[col])
            le_count += 1
print('{} colunas foram codificadas por rótulo.'.format(le_count))
```

```python
# converter o resto da variável categórica em dummy
df_HR = pd.get_dummies(df_HR, drop_first=True)
```

```python
print(df_HR.shape)
df_HR.head()
```

### Escalonamento de Recursos

O escalonamento de recursos usando MinMaxScaler essencialmente reduz o intervalo de forma que o intervalo esteja agora entre 0 e n. 

Os algoritmos de aprendizado de máquina têm melhor desempenho quando as variáveis ​​numéricas de entrada se enquadram em uma escala semelhante. Nesse caso, estamos escalando entre 0 e 5.

```python
# importando o MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 5))
HR_col = list(df_HR.columns)
HR_col.remove('Attrition')
for col in HR_col:
    df_HR[col] = df_HR[col].astype(float)
    df_HR[[col]] = scaler.fit_transform(df_HR[[col]])
df_HR['Attrition'] = pd.to_numeric(df_HR['Attrition'], downcast='float')
df_HR.head()
```

```python
print('Tamanho do conjunto de dados totalmente codificado: {}'. format(df_HR.shape))
```

## Construindo Modelos de Aprendizado de Máquina

## Algoritmos baseline 

Vamos primeiro usar uma variedade de algoritmos como baseline (usando hiperparâmetros prontos para uso) antes de passarmos para soluções mais sofisticadas. Os algoritmos considerados nesta seção são: Logistic Regression, Random Forest, SVM, KNN, Decision Tree Classifier, Gaussian NB.

Vamos avaliar cada modelo separadamente e fornecer pontuações de precisão e desvio padrão.

```python
# seleção de algoritmos para considerar e definir a medida de desempenho
models = []
models.append(('Logistic Regression', LogisticRegression(solver='liblinear', random_state=7,
                                                         class_weight='balanced')))
models.append(('Random Forest', RandomForestClassifier(
    n_estimators=100, random_state=7)))
models.append(('SVM', SVC(gamma='auto', random_state=7)))
models.append(('KNN', KNeighborsClassifier()))
models.append(('Decision Tree Classifier',
               DecisionTreeClassifier(random_state=7)))
models.append(('Gaussian NB', GaussianNB()))
```

Vamos avaliar cada modelo por vez e fornecer pontuações de precisão e desvio padrão

```python
acc_results = []
auc_results = []
names = []
# definir tabela em tabela para preencher os resultados de desempenho
col = ['Algorithm', 'ROC AUC Mean', 'ROC AUC STD', 
       'Accuracy Mean', 'Accuracy STD']
df_results = pd.DataFrame(columns=col)
i = 0
# avaliar cada modelo usando validação cruzada
for name, model in models:
    kfold = model_selection.KFold(
        n_splits=10, shuffle=False)  # 10-fold cross-validation

    cv_acc_results = model_selection.cross_val_score(  # accuracy scoring
        model, X_train, y_train, cv=kfold, scoring='accuracy')

    cv_auc_results = model_selection.cross_val_score(  # roc_auc scoring
        model, X_train, y_train, cv=kfold, scoring='roc_auc')

    acc_results.append(cv_acc_results)
    auc_results.append(cv_auc_results)
    names.append(name)
    df_results.loc[i] = [name,
                         round(cv_auc_results.mean()*100, 2),
                         round(cv_auc_results.std()*100, 2),
                         round(cv_acc_results.mean()*100, 2),
                         round(cv_acc_results.std()*100, 2)
                         ]
    i += 1
df_results.sort_values(by=['ROC AUC Mean'], ascending=False)
```

```python
fig = plt.figure(figsize=(15, 7))
fig.suptitle('Comparação de precisão de algoritmo')
ax = fig.add_subplot(111)
plt.boxplot(acc_results)
ax.set_xticklabels(names)
plt.show()
```

A precisão da classificação é o número de previsões corretas feitas como uma proporção de todas as previsões feitas. É a métrica de avaliação mais comum para problemas de classificação.


No entanto, muitas vezes é mal utilizado , pois só é realmente adequado quando há um número igual de observações em cada classe e todas as previsões e erros de previsão são igualmente importantes.

 Não é o caso neste projeto, portanto, uma métrica de pontuação diferente pode ser mais adequada.

A área sob a curva ROC (ou AUC para abreviar) é uma métrica de desempenho para problemas de classificação binária. 

O AUC representa a capacidade de um modelo para discriminar entre classes positivas e negativas e é mais adequado para este projeto. Uma área de 1,0 representa um modelo que fez todas as previsões perfeitamente. Uma área de 0,5 representa um modelo tão bom quanto aleatório.

```python
fig = plt.figure(figsize=(15, 7))
fig.suptitle('Comparação do algoritmo ROC AUC')
ax = fig.add_subplot(111)
plt.boxplot(auc_results)
ax.set_xticklabels(names)
plt.show()
```

Com base em nossa análise de comparação ROC AUC, a regressão logística e a floresta aleatória mostram as pontuações médias mais altas da AUC. Vamos listar esses dois algoritmos para uma análise posterior. 

#### **Regressão Logística**

A regressão logística é um algoritmo de classificação de aprendizado de máquina usado para prever a probabilidade de uma variável dependente categórica. A regressão logística é um algoritmo de classificação que não é tão sofisticado quanto os métodos de conjunto ou o método de árvores de decisão, porém, ela nos fornece uma boa referência.

#### **Random Forest**

Random Forest é um método popular e versátil de aprendizado de máquina, capaz de resolver regressões e classificações. Random Forest é uma marca de aprendizagem do Ensemble, pois depende de um conjunto de árvores de decisão. Agrega Árvores de Classificação (ou Regressão). Uma árvore de decisão é composta por uma série de decisões que podem ser usadas para classificar uma observação em um conjunto de dados.

A Random Forest ajusta vários classificadores de árvore de decisão em várias subamostras do conjunto de dados e usa a média para melhorar a precisão preditiva e o sobreajuste de controle. Random Forest pode lidar com um grande número de recursos e é útil para estimar quais de suas variáveis são importantes nos dados subjacentes que estão sendo modelados.


## **Logistic Regression**

Vamos examinar mais de perto o uso do algoritmo de regressão logística. Usarei a validação cruzada de 10 vezes para treinar nosso modelo de regressão logística e estimar sua pontuação AUC.

### Fine-tuning

O GridSearchCV permite o uso para ajustar hiperparâmetros pesquisando valores de parâmetros especificados para um estimador. Conforme mostrado abaixo, os resultados do GridSearchCV nos forneceram um hiperparâmetro ajustado usando ROC_AUC como a métrica de pontuação.

```python
kfold = model_selection.KFold(n_splits=10,  shuffle=False)
modelCV = LogisticRegression(solver='liblinear',
                             class_weight="balanced", 
                             random_state=7)
scoring = 'roc_auc'
results = model_selection.cross_val_score(
    modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("AUC score (STD): %.2f (%.2f)" % (results.mean(), results.std()))
```

```python
param_grid = {'C': np.arange(1e-03, 2, 0.01)} # lista de hiperparâmetros para ajuste fino
log_gs = GridSearchCV(LogisticRegression(solver='liblinear', # configurando GridSearchCV
                                         class_weight="balanced", 
                                         random_state=7),
                      #iid=True,
                      return_train_score=True,
                      param_grid=param_grid,
                      scoring='roc_auc',
                      cv=10)

log_grid = log_gs.fit(X_train, y_train)
log_opt = log_grid.best_estimator_
results = log_gs.cv_results_

print('='*20)
print("Melhor parametro: " + str(log_gs.best_estimator_))
print("Melhor parametro: " + str(log_gs.best_params_))
print('Melhor score:', log_gs.best_score_)
print('='*20)
```

Conforme mostrado acima, os resultados do GridSearchCV nos forneceram um hiperparâmetro ajustado usando ROC_AUC como a métrica de pontuação.

### Matriz de confusão

## **Gráficos ROC**

A curva AUC - ROC é uma medida de desempenho para problemas de classificação em várias configurações de limites. 

ROC é uma curva de probabilidade e AUC representa o grau ou medida de separabilidade. Diz o quanto o modelo é capaz de distinguir entre as classes. 

A linha verde representa a curva ROC de um classificador puramente aleatório; um bom classificador fica o mais longe possível dessa linha (em direção ao canto superior esquerdo).

```python
# Create ROC Graph
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, log_opt.predict_proba(X_test)[:,1])
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf_opt.predict_proba(X_test)[:,1])
plt.figure(figsize=(14, 6))

# Plot Logistic Regression ROC
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
# Plot Random Forest ROC
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_opt_roc_auc)
# Plot Base Rate ROC
plt.plot([0,1], [0,1],label='Base Rate' 'k--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de falso positivo')
plt.ylabel('Taxa de verdadeiro positivo')
plt.title('Gráfico ROC')
plt.legend(loc="lower right")
plt.show()
```

Conforme mostrado acima, o modelo de Regressão Logística ajustada mostrou uma pontuação AUC mais alta em comparação com o Classificador Random Forest.

### **Indicadores e Plano Estratégico de Retenção**

Os indicadores mais fortes de saída de pessoas incluem:

* Renda Mensal : pessoas com salários mais altos têm menor probabilidade de deixar a empresa. Portanto, esforços devem ser feitos para reunir informações sobre os benchmarks da indústria no mercado local atual para determinar se a empresa está oferecendo salários competitivos.

* Horas extras: as pessoas que trabalham horas extras têm maior probabilidade de deixar a empresa. Conseqüentemente, esforços devem ser feitos para definir o escopo adequado dos projetos antecipadamente, com suporte e mão de obra adequados, de modo a reduzir o uso de horas extras.

* Idade : os funcionários com idade relativamente jovem entre 25 e 35 anos têm maior probabilidade de sair. Assim, esforços devem ser feitos para articular claramente a visão de longo prazo da empresa e os jovens funcionários se enquadram nessa visão, bem como fornecer incentivos na forma de caminhos claros de promoção, por exemplo.

* DistanceFromHome : Funcionários que moram longe de casa têm maior probabilidade de deixar a empresa. Dessa forma, esforços devem ser feitos no sentido de apoiar na forma de transporte empresarial para aglomerados de funcionários que saem da mesma área, ou na forma de Auxílio Transporte. A triagem inicial de funcionários com base em sua localização residencial provavelmente não é recomendada, pois seria considerada uma forma de discriminação, desde que os funcionários cheguem ao trabalho pontualmente todos os dias.

* TotalWorkingYears : Os funcionários mais experientes têm menos probabilidade de sair. Os funcionários que têm entre 5 e 8 anos de experiência devem ser identificados como tendo um risco maior de demissão.

* YearsAtCompany : As empresas leais têm menos probabilidade de sair. Os funcionários que completam seu aniversário de dois anos devem ser identificados como tendo um risco potencialmente maior de sair.

* YearsWithCurrManager : Um grande número de pessoas que saem da empresa 6 meses depois de seus gerentes atuais. Usando os detalhes do gerente para cada funcionário, pode-se determinar qual gerente teve o maior número de funcionários demitidos no ano passado.

Várias métricas podem ser usadas aqui para determinar se uma ação deve ser realizada com um gerente:

* Número de funcionários sob gestão com altas taxas de rotatividade: isso indicaria que a estrutura da organização pode precisar ser revisada para melhorar a eficiência

* Número de anos que o Gerente está em uma posição específica: isso pode indicar que os funcionários podem precisar de treinamento de gestão ou ser designado a um mentor (de preferência um Executivo) na organização

* Padrões nos funcionários que pediram demissão: isso pode indicar padrões recorrentes na saída dos funcionários, caso em que ações possam ser tomadas em conformidade.
