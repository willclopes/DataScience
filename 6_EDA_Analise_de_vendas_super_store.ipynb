## Contexto: O que √© an√°lise de vendas?

An√°lise de vendas √© a pr√°tica de gerar insights de dados de vendas, tend√™ncias e m√©tricas para definir metas e prever o desempenho de vendas futuro. A pr√°tica recomendada para an√°lise de vendas √© vincular estreitamente todas as atividades para determinar os resultados da receita e definir objetivos para sua equipe de vendas.

A an√°lise deve se concentrar na melhoria e no desenvolvimento de uma estrat√©gia para melhorar o desempenho de vendas a curto e longo prazo. Um exemplo comum de atividade de an√°lise de vendas √© definir objetivos espec√≠ficos de fun√ß√£o para sua equipe na forma de KPIs (*Key Performance Indicator*) ou m√©tricas. Por exemplo, definir uma meta de receita para seu diretor de vendas e, ao mesmo tempo, definir uma meta de produtividade de vendas para sua equipe de gerenciamento de contas.

## Por que monitorar a an√°lise de vendas?

Se o seu neg√≥cio fosse uma banda de rock, as vendas seriam o vocalista. Todos os olhos est√£o voltados para o desempenho de vendas e receita, e um desempenho forte (ou fraco) pode se tornar uma for√ßa de mobiliza√ß√£o potente para toda a equipe.

O monitoramento da an√°lise de vendas na forma de m√©tricas de vendas ajuda a aumentar seu desempenho, otimizar as atividades de vendas e melhorar a responsabilidade. Sua equipe de vendas tem uma ampla gama de atividades para se concentrar e operar em um ambiente de ritmo acelerado. Uma estrat√©gia de an√°lise de vendas bem definida fornece √† sua equipe foco e clareza para que eles possam se concentrar em fazer o que fazem melhor.


## O que √© a EDA

A EDA ou an√°lise explorat√≥ria de dados √© uma etapa importante do aprendizado de m√°quina que envolve aprender sobre os dados sem gastar muito tempo ou se perder nisso. Nesse processo voc√™ se familiariza com a estrutura e as caracter√≠sticas gerais do conjunto de dados, com os recursos independentes e dependentes e suas intera√ß√µes. Voc√™ tamb√©m ter√° uma ideia dos processos de limpeza de dados a serem empregados e das poss√≠veis t√©cnicas de modelagem.

## An√°lise explorat√≥ria de dados

### 1. Qual categoria √© a mais vendida e mais lucrativa?

(Mais vendido significa olhar para as vendas e ver qual categoria est√° sendo mais lucrativa em vendas. Vamos agrupar Vendas, Lucro e tamb√©m Quantidade por Categoria e executar as fun√ß√µes df.group (). Sum e pd.DataFrame ().)

```python
# Primeiro vamos agrupar por categoria
category_analysis = pd.DataFrame(superstore.groupby(['Category'])[['Sales', 'Profit', 'Quantity']].sum())
category_analysis
```

```python
# difinir o formato do gr√°fico
sns.set_theme(style="whitegrid")
figure, axis = plt.subplots(1, 3, figsize=(8, 5))
# critar os plots
cat1 = sns.barplot(x = category_analysis.index, y = category_analysis.Sales, ax=axis[0])
cat2 = sns.barplot(x = category_analysis.index, y = category_analysis.Profit, ax=axis[1])
cat3 = sns.barplot(x = category_analysis.index, y = category_analysis.Quantity, ax=axis[2])
# definir os titulos
cat1.set(title = 'Vendas')
cat2.set(title = 'Lucro')
cat3.set(title = 'Quantidade')
# rotacionar os eixos
plt.setp(cat1.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat2.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat3.get_xticklabels(), rotation = 'vertical', size = 9)
# definir o espa√ßamento
figure.tight_layout()
```

As observa√ß√µes s√£o:

* Todas as 3 categorias - M√≥veis, Material de Escrit√≥rio e Tecnologia t√™m vendas semelhantes.

* Tecnologia √© a mais vendida e √© bom saber que essa categoria tamb√©m √© a mais lucrativa. Apenas uma quantidade m√≠nima √© vendida, pois esses produtos geralmente s√£o compras √∫nicas que podem durar pelo menos 4-5 anos.

* Embora M√≥veis fa√ßa vendas semelhantes √†s de Tecnologia, √© o menos lucrativo e a quantidade vendida tamb√©m √© m√≠nima.

* O material de escrit√≥rio vende mais em termos de quantidade, pois √© um produto relativamente barato.

### 2. Quais s√£o as subcategorias mais vendidas e mais lucrativas?

```python
# agrupar por sub-categoria
subcat_analysis = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Sales', 'Profit']].sum())
# ordernar de acordo com "Sales"
subcat_sales = pd.DataFrame(subcat_analysis.sort_values('Sales', ascending = False))
subcat_sales
```

```python
# ordernar de acordo com "Profit"
subcat_profit = pd.DataFrame(subcat_analysis.sort_values('Profit', ascending = False))
subcat_profit
```

```python
# Gerar os gr√°fico
sns.set_theme(style="whitegrid")

figure, axis = plt.subplots(1, 2, figsize=(12, 6))

subcat1 = sns.barplot(data = subcat_sales, x = subcat_sales.index, y = subcat_sales.Sales, ax=axis[0])
subcat1.set(title="Sub-Categoria mais vendida")
subcat1.set_xticklabels(subcat1.get_xticklabels(),rotation = "vertical", size = 10)

subcat2 = sns.barplot(data = subcat_profit, x = subcat_profit.index, y = subcat_profit.Profit, ax=axis[1])
subcat2.set(title = "Sub-Categoria mais lucrativa")
subcat2.set_xticklabels(subcat2.get_xticklabels(),rotation = "vertical", size = 10)

figure.tight_layout()
plt.show()
```

Vamos analisar os gr√°ficos de barra:

* Telefones e cadeiras s√£o as 2 subcategorias mais vendidas.

* As copiadoras geram mais lucro, seguidas por telefones, acess√≥rios, pap√©is e pastas. A estrat√©gia de marketing deve se concentrar no marketing desses produtos.

* Na outra ponta do espectro, m√°quinas, fixadores, suprimentos, estantes e mesas t√™m uma margem de perdas pr√≥xima de zero. Esses s√£o produtos que a 

* A Super Store pode considerar retirar do cat√°logo de produtos ou aumentar o pre√ßo de venda e a margem de lucro ou negociar por um pre√ßo mais baixo com o fornecedor.

### 3. Qual √© a subcategoria mais vendida?

```python
subcat_quantity = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Quantity']].sum().sort_values('Quantity',ascending=False))
subcat_quantity
```

```python
# gr√°fico
sns.set_theme(style="whitegrid")
sns.barplot(data = subcat_quantity, y = subcat_quantity.index, x = subcat_quantity.Quantity, palette = "muted")
plt.title("Sub-Category mais lucrativa")
plt.show()
```

Aqui, podemos deduzir que

* A super loja deve garantir que o estoque esteja sempre bem abastecido para a subcategoria mais vendida, como fich√°rios, papel, m√≥veis e telefones.

* Apesar de ser mais rent√°vel, as copiadoras vendem menos apenas 234, mas como se trata de um equipamento de escrit√≥rio relativamente caro e costuma ser usado por poucos anos, √© compreens√≠vel que vende menos entre todos.

### 4. Qual segmento de cliente √© mais lucrativo?

```python
segment_analysis = pd.DataFrame(superstore.groupby(['Segment'])[['Profit']].sum())
segment_analysis
```

```python
# gr√°fico
sns.set_theme(style="whitegrid")
sns.barplot(data = segment_analysis, x = segment_analysis.index, y = segment_analysis.Profit, palette = "rocket")
plt.title("Segmento de clientes mais lucrativo")
plt.show()
```

O segmento de consumo √© o mais lucrativo, seguido pelo Segmento Corporativo e Home Office. Conseq√ºentemente, a estrat√©gia de marketing deve ter como alvo ou enfocar mais a reten√ß√£o de clientes do segmento corporativo e de consumo.

### 5. Qual √© o modo de envio preferido?

```python
sns.set_theme(style="whitegrid")
sns.countplot(superstore['Ship Mode'])
plt.title("Modo de Envio")
plt.show()
```

A "Classe Padr√£o" √© o m√©todo de envio preferido e talvez o mais barato tamb√©m. Os outros modos n√£o s√£o populares entre os clientes e podem ser muito caros.

### 6. Qual regi√£o √© a mais lucrativa?

```python
region_analysis = pd.DataFrame(superstore.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
```

```python
# gr√°fico
explode = [0, 0, 0, 0.1]
plt.pie(region_analysis.Profit, labels = region_analysis.Region, startangle = 90, autopct = "%1.0f%%", explode = explode, shadow = True)
plt.title("Mais lucros por regi√£o")
plt.show()
```

As regi√µes Leste e Oeste s√£o as mais lucrativas.

### 7. Qual cidade tem o maior n√∫mero de vendas?

```python
city_sales = pd.DataFrame(superstore.groupby(['City'])['Sales', 'Quantity'].sum().sort_values('Sales',ascending = False))
top10 = city_sales[:10]
top10
```

```python
bottom10 = city_sales[-10:]
bottom10
```

```python
# gr√°fico
figure, axis = plt.subplots(1, 2, figsize=(12, 5))
sns.set_theme(style="whitegrid")
top10c = sns.barplot(data = top10, y = top10.index, x = top10.Sales, palette = "coolwarm", ax = axis[0])
top10c.set(Title = "As 10 cidades com vendas mais altas")
top10c.set_yticklabels(top10c.get_yticklabels(),size = 10)

bottom10c = sns.barplot(data = bottom10, y = bottom10.index, x = bottom10.Sales, palette = "coolwarm", ax=axis[1])
bottom10c.set(Title = "As 10 cidades com vendas mais baixas")
bottom10c.set_yticklabels(bottom10c.get_yticklabels(),size = 10)

figure.tight_layout()
plt.show()
```

H√° uma grande disparidade entre as cidades com vendas mais altas e mais baixas. A estrat√©gia de marketing deve ter como alvo as 10 principais cidades.
