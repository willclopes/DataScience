# **Introdução**

A fraude de cartão de crédito causa bilhões de reais em danos a cada ano. Os casos mais infames afetaram dezenas a centenas de milhões de consumidores em ataques únicos por meio da exposição ilegal de informações de identificação pessoal (PII) relacionadas a cartões de crédito. 

Casos isolados também são comuns e podem ser causados ​​por uma variedade de métodos, incluindo skimming, engenharia social e fraude de aplicativo.

Para proteger seus clientes, as empresas de cartão de crédito contam com software de detecção e prevenção de fraude para analisar compras com cartão de crédito. Esses programas procuram padrões incomuns ou inesperados para classificá-los como possivelmente fraudulentos.




## **Contexto:**

A **detecção de fraude** é um conjunto de atividades realizadas para evitar que dinheiro ou propriedade sejam obtidos por meio de falsos pretextos.
A fraude pode ser cometida de diferentes maneiras e em diferentes ambientes. Por exemplo, a fraude pode ser cometida nos setores de bancos, seguros, governo e saúde.


Um tipo comum de fraude bancária é a apropriação da conta do cliente. Isso ocorre quando alguém obtém acesso ilegal à conta bancária da vítima usando bots . Outros exemplos de fraude no setor bancário incluem o uso de aplicativos maliciosos, o uso de identidades falsas, lavagem de dinheiro , fraude de cartão de crédito e fraude móvel. Fraude de seguro inclui fraude de desvio de prêmio, que é o desvio de prêmios de seguro, ou rotatividade gratuita, que é negociação excessiva por um corretor da bolsa para maximizar as comissões. 

Outras formas de fraude de seguro incluem desvio de ativos, fraude de compensação de trabalhadores, fraude de acidente de carro, fraude de carro roubado ou danificado e fraude contra incêndio domiciliar. 

O motivo por trás de todas as fraudes de seguro são os lucros financeiros.
Fraude governamental consiste em cometer fraude contra agências federais, como o Departamento de Saúde e Serviços Humanos dos Estados Unidos, o Departamento de Transporte, o Departamento de Educação ou o Departamento de Energia. Os tipos de fraude governamental incluem cobrança de procedimentos desnecessários, cobrança excessiva de itens que custam menos, fornecimento de equipamento antigo ao cobrar por equipamento novo e relatórios de horas trabalhadas para um trabalhador que não existe.

Fraude de saúde inclui fraude de drogas e fraude médica, bem como fraude de seguro. A fraude na área de saúde é cometida quando alguém defrauda uma seguradora ou programa de saúde do governo.

A fraude pode ser cometida de diferentes maneiras e em muitos setores. A maioria dos métodos de detecção **combina uma variedade de conjuntos de dados** para formar uma visão geral conectada de dados de pagamento válidos e não válidos para tomar uma decisão. Essa decisão deve levar em consideração o endereço IP, geolocalização, identificação do dispositivo, dados “BIN”, latitude / longitude global, padrões históricos de transação e as informações reais da transação. Na prática, isso significa que os comerciantes e emissores implementam respostas com base analítica que usam dados internos e externos para aplicar um conjunto de regras de negócios ou algoritmos analíticos para detectar fraudes.

A Detecção de Fraudes em Cartões de Crédito com Aprendizado de Máquina é um processo de investigação de dados por uma equipe de Data Science e o desenvolvimento de um modelo que proporcionará os melhores resultados na revelação e prevenção de transações fraudulentas. A informação é então executada através de um modelo sutilmente treinado que encontra padrões e regras para que possa classificar se uma transação é fraudulenta ou legítima.







## **Declaração do problema**

Cada vez mais empresas e consumidores estão migrando para formas de pagamento online. Milhões de pessoas são vítimas de fraudes de cartão de crédito que acabam custando milhares de reais. O Wall Street Journal relatou que as perdas por fraude, incluindo perdas vinculadas a cartões de crédito e débito, custaram a bancos, comerciantes e, em alguns casos, titulares de cartões de $ 16,9 bilhões de dólares em 2019. 

Mais recentemente, o Relatório Nilson quantificou as perdas globais por fraude de cartão em $ 28,65 bilhões de dólares.

## Preparação de dados

Uma vez que não temos variáveis ​​categóricas nos dados do conjunto de dados, não precisamos convertê-los em variáveis "dummys" antes de usar nossos dados para modelagem

```python
df.columns.values
```

A variável de resultado é “fraude” e todas as outras variáveis ​​são preditivas, portanto, vamos separá-las.

```python
df_vars=df.columns.values.tolist()
y=['fraud']
X=[i for i in df_vars if i not in y]
```

## Feature Selection

O [Recursive Feature Elimination (RFE)](http://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.RFE.html) funciona removendo variáveis ​​recursivamente e construindo um modelo sobre as variáveis ​​que permanecem. Ele usa a precisão do modelo para identificar quais variáveis ​​(e combinação de variáveis) contribuem mais para prever o atributo de destino.
Vamos usar a seleção de recursos para nos ajudar a decidir quais variáveis ​​são significativas e podem prever transações fraudulentas com grande precisão.

```python
len (X)
```

Há um total de 29 colunas em X, que tal selecionar 10?

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
rfe = RFE(estimator=model, n_features_to_select= 10)
rfe = rfe.fit(df[X], df[y])
print(rfe.support_)
print(rfe.ranking_)
```

```python
print(rfe.support_)
```

```python
print(rfe.ranking_)
```

Vemos que o RFE escolheu as 10 variáveis ​​para nós, que são marcadas como True na matriz support_ e marcadas com uma opção “1” na matriz ranking_ . Vamos encontrá-los.

```python
data_X1 = pd.DataFrame({
 'Feature': df[X].columns,
 'Importance': rfe.ranking_},)
data_X1.sort_values(by=['Importance'])
```

```python
cols=[]
for i in range (0, len(data_X1["Importance"])):
    if data_X1["Importance"][i] == 1:
        cols.append(data_X1["Feature"][i])
print(cols)
print(len(cols))
```

## Modelagem

A fraude geralmente envolve vários métodos repetidos, tornando a busca de padrões um foco geral para a detecção de fraudes. Por exemplo, os analistas de dados podem evitar fraudes em seguros criando algoritmos para detectar padrões e anomalias.

A detecção de fraudes pode ser separada pelo uso de técnicas de análise de dados estatísticos ou inteligência artificial ( IA ).

As técnicas de análise de dados estatísticos incluem:

* cálculo de parâmetros estatísticos
* análise de regressão
* distribuições e modelos de probabilidade
* correspondência de dados

As técnicas de IA usadas para detectar fraude incluem:

* A mineração de dados classifica, agrupa e segmenta dados para pesquisar milhões de transações para encontrar padrões e detectar fraudes.

* As redes neurais aprendem padrões de aparência suspeita e usam esses padrões para detectá-los melhor.

* O aprendizado de máquina identifica automaticamente as características encontradas na fraude.

* O reconhecimento de padrões detecta classes, clusters e padrões de comportamento suspeito.

```python
X=df[cols]
y=df['fraud']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Logistic Regression

```python

logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, y_train)
```

```python
print('Acurácia do Logistic regression: {:.3f}'.format(accuracy_score(y_test, logreg.predict(X_test))))
```

A validação cruzada tenta evitar overfitting enquanto ainda produz uma previsão para cada conjunto de dados de observação. Estamos usando a validação cruzada de 10 vezes para treinar nosso modelo de regressão logística.

```python
kfold = model_selection.KFold(n_splits=10, random_state=42, shuffle=True)
modelCV = LogisticRegression(random_state=42)
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("Precisão média da validação cruzada de 10 vezes do modelo de regressão logística: %.3f" % (results.mean()))
```

**Precisão média da validação cruzada de 10 vezes do modelo de regressão logística: 0,983**

A precisão média permanece muito próxima da precisão do modelo de regressão logística; portanto, podemos concluir que o modelo generaliza bem.

### Random Forest

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

```python
print('Acurácia com Random Forest: {:.3f}'.format(accuracy_score(y_test, rf.predict(X_test))))
```

```python
kfold = model_selection.KFold(n_splits=10, random_state=42, shuffle=True)
modelCV =RandomForestClassifier(random_state=42)
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("Precisão média da validação cruzada de 10 vezes do modelo com o modelo com random forest: %.3f" % (results.mean()))
```

**Precisão média da validação cruzada de 10 vezes do modelo de floresta aleatório: 0,986**

A precisão média permanece muito próxima da precisão do modelo Random Forest; portanto, podemos concluir que o modelo generaliza bem.


### Support Vector Machine

```python
from sklearn.svm import SVC
svc = SVC(random_state=42)
svc.fit(X_train, y_train) 
```

```python
print('Acuracia com Support vector machine: {:.3f}'.format(accuracy_score(y_test, svc.predict(X_test))))
```

```python
kfold = model_selection.KFold(n_splits=10, random_state=42, shuffle=True)
modelCV = SVC(random_state=42)
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("Precisão média da validação cruzada de 10 vezes do modelo com SVM: %.3f" % (results.mean()))
```

**Precisão média da validação cruzada de 10 vezes do modelo de máquina de vetor de suporte: 0,983**

A precisão média permanece muito próxima da precisão do modelo da Máquina de Vetores de Suporte; portanto, podemos concluir que o modelo generaliza bem.

## Conclusão

O vencedor é ... **Random Forest**.

Quando uma transação era fraudulenta, com que frequência o classificador prevê isso corretamente? Essa medição é chamada de “recall” e uma rápida olhada nesses diagramas pode demonstrar que a random forest é claramente a melhor para este critério. 

De todos os casos fraudulentos, o random forest recuperou corretamente 1142 de 1189. Isso se traduz em um "recall" de volume de negócios de cerca de 96% (1142/1189), melhor do que a regressão logística (95%) (1128/1189) ou vetor de suporte máquina (94%) 1123/1189).


Quando um classificador prevê que uma transação será fraudulenta, com que frequência essa transação é realmente fraudulenta? Esta medição é chamada de “precisão”. O random forest tem cerca de 99% de precisão (1142 de 1149) com regressão logística em cerca de 99% (1128 de 1141) e máquina de vetor de suporte em cerca de 100% (1123 de 1127).

### A Curva ROC

A curva de característica de operação do receptor (ROC) é outra ferramenta comum usada com classificadores binários. A linha pontilhada representa a curva ROC de um classificador puramente aleatório; um bom classificador fica o mais longe possível dessa linha (em direção ao canto superior esquerdo).

```python
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
rf_roc_auc = roc_auc_score(y_test, rf.predict(X_test))
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc, color='blue')
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_roc_auc, color='green')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falso Positivo')
plt.ylabel('Taxa de verdadeiro positivo')
plt.title('Curva ROC')
#plt.legend(loc="lower right")
#plt.savefig('ROC')
plt.show()
```

### Importância do recurso para o modelo de random forest


A importância do recurso avalia a importância de cada recurso para a decisão que uma árvore toma. É um número entre 0 e 1 para cada recurso, onde 0 significa “nem um pouco usado” e 1 significa “prediz perfeitamente o destino”. As importâncias do recurso sempre somam 1.

```python
feature_labels = X.columns
importance = rf.feature_importances_
feature_indexes_by_importance = importance.argsort()[::-1]
for index in feature_indexes_by_importance:
    print('{}-{:.2f}%'.format(feature_labels[index], (importance[index] *100.0)))
```

De acordo com nosso modelo Random Forest, o acima mostra as características mais importantes que influenciam se uma transação será fraudulenta, em ordem decrescente.
Então, podemos visualizar as importâncias do recurso:

```python
df_features = [x for i,x in enumerate(X.columns) if i!=10]

sns.set(style="white")
matplotlib.rcParams['figure.figsize'] = (14.0, 7.0)
def plot_feature_importances(model):
    plt.figure(figsize=(8,6))
    n_features = 10
    plt.barh(range(n_features), rf.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), df_features)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
plot_feature_importances(rf)
#plt.savefig('feature_importance')
```

O recurso “N” é de longe o recurso mais importante.
