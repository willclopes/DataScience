# Introdução

Você pode prever os clientes que provavelmente deixarão a empresa?

Ao escolher um provedor de serviços de telefonia, os clientes geralmente têm muitas opções. Eles podem escolher qualquer provedor de serviços e podem se afastar do provedor atual. 

Quando um cliente opta por mudar do provedor atual para um novo provedor, isso resulta em perda de negócios e receita para o provedor atual. **O percentual de clientes que saem e desconectam do serviço é conhecido como “churn”.** 


**Uma base de clientes estável é a chave para o sucesso de qualquer negócio.**

As empresas tentam manter os clientes satisfeitos, para retê-los o maior tempo possível. No entanto, no mundo real, a rotatividade de clientes pode chegar a 25% ao ano no setor de telecomunicações. Além disso, o custo de adquirir um novo cliente é 10 vezes maior do que o custo de reter um cliente existente. Isso representa um sério desafio para os proprietários de empresas.



## Contexto: CHURN

A rotatividade de clientes é uma das métricas mais importantes para uma empresa em crescimento avaliar. Embora não seja a medida mais animadora, é um número que pode dar à sua empresa a dura verdade sobre a retenção de clientes.

É difícil medir o sucesso se você não medir as falhas inevitáveis ​​também. Embora você se esforce para que 100% dos clientes fiquem com sua empresa, isso é simplesmente irreal. É aí que entra a rotatividade de clientes.

A rotatividade de clientes é a porcentagem de clientes que pararam de usar o produto ou serviço de sua empresa durante um determinado período. Você pode calcular a taxa de rotatividade dividindo o número de clientes perdidos durante aquele período - digamos um trimestre - pelo número de clientes que você tinha no início do período.

Por exemplo, se você iniciar o trimestre com 400 clientes e terminar com 380, sua taxa de desligamento será de 5% porque você perdeu 5% dos clientes.

Obviamente, sua empresa deve almejar uma taxa de churn o mais próximo possível de 0%. Para fazer isso, sua empresa deve estar sempre no topo da taxa de rotatividade e tratá-la como uma prioridade.

### Por que a taxa de rotatividade do cliente é importante?

Você pode estar se perguntando por que é necessário calcular a taxa de rotatividade. Naturalmente, você vai perder alguns clientes aqui e ali, e 5% não parece tão ruim, certo?

Bem, é importante porque custa mais adquirir novos clientes do que reter os clientes existentes. Na verdade, um aumento de apenas 5% na retenção de clientes pode gerar um aumento de pelo menos 25% no lucro. Isso ocorre porque os clientes recorrentes provavelmente gastarão 67% a mais nos produtos e serviços da sua empresa. Como resultado, sua empresa pode gastar menos com os custos operacionais de ter que adquirir novos clientes. Você não precisa gastar tempo e dinheiro para convencer um cliente existente a escolher sua empresa em vez dos concorrentes, porque eles já tomaram essa decisão.

Novamente, pode parecer que uma taxa de rotatividade de 5% é sólida e saudável. Você ainda pode ter uma grande receita com essa taxa de rotatividade.

### Por que os clientes mudam?

* **O cliente não quer mais seu produto.**

O que quer que tenha atraído o cliente em primeiro lugar, acabou. Quer sejam os valores da empresa, a qualidade do produto ou o preço em que o produto foi apresentado pela primeira vez.

Isso pode ser uma atualização de IU incorreta, falta de recursos importantes ou até mesmo algo tão simples como um pequeno bug.

* **O cliente encontrou uma alternativa melhor**

Este é provavelmente o mais comum, pois os clientes estão sempre em busca de um negócio melhor.

## **Logistic Regression**

Em seguida, procedemos à criação do modelo de Regressão Logística

Com base na declaração do problema, precisamos de um modelo preditivo que possa fazer uma classificação binária ou prever o tipo Sim / Não ou 1/0 de variável de saída. 

Um modelo preditivo comumente implementado para classificação binária e previsão do resultado binário é a regressão logística. 

A regressão logística é um algoritmo de **classificação binária** pertencente ao modelo de regressão linear generalizado. Também pode ser usado para resolver problemas com mais de 2 classes. 

É possível usar a regressão logística para criar um modelo usando os dados de rotatividade do cliente e usá-lo para prever se um determinado cliente de um conjunto de clientes descontinuará o serviço.

Por exemplo, uma das variáveis ​​nos dados pode ser a “receita anual”. Outra variável é o “gênero” do cliente. O resultado da função de regressão logística nos dirá como a renda e / ou sexo determinam a probabilidade de interrupção do serviço pelo cliente.


### **Equação de regressão logística e função sigmóide**

A seguir está a função de regressão logística:

**P (Y = 1 | X) = 1 / (1+ e ^ (- 1x (β0 + β1X1 + β2X2 + β3X3 + β4X4 + ⋯ + βnXn)))**

onde:
β0 a βn são vários coeficientes
X0 a Xn são as variáveis ​​independentes que impactam a variável dependente
E P (Y = 1 | X) é a probabilidade de um resultado positivo.

Observe o expoente na função. É aqui que a regressão linear atua em (β0 + β1X1 + β2X2 + β3X3 + β4X4 + ⋯ + βnXn).

A função de regressão logística é uma função sigmóide. Essa função possui valores de output entre 0 e 1 com transição entre os níveis. Essa característica da função ajuda na previsão de resultados binários. Com base no valor das variáveis, a saída pode estar no nível 1 ou 0, que corresponde à probabilidade de o cliente sair da empresa ou continuar com ela.

### **Premissas para regressão logística**
As seguintes suposições são feitas para a regressão logística:

1. A regressão logística binária requer que a variável dependente seja binária e siga uma distribuição binomial (por exemplo, um cliente descontinuará o serviço ou não, Sim ou Não). Para mais de 2 resultados (ordinal), a regressão logística requer que as categorias da variável dependente sejam mutuamente exclusivas e exaustivas.

2. As observações devem ser independentes umas das outras (por exemplo, os dados de um cliente não devem depender dos dados de outro cliente, ou o mesmo cliente não deve ser usado repetidamente nos dados)

3. Não deve haver multicolinearidade entre as variáveis ​​independentes (por exemplo, evite usar variáveis ​​dos dados do cliente que dependem umas das outras, digamos, cidade, estado, município e código postal não são todos independentes)

4. A linearidade das variáveis ​​independentes em relação às probabilidades de log da variável dependente (por exemplo, as probabilidades de log da probabilidade de o cliente descontinuar o serviço devem estar linearmente relacionadas a várias variáveis ​​como sexo, renda, etc.)

5. Grande tamanho de amostra (por exemplo, dados com muitos registros)


### Variáveis Dummy e Normalização

Os campos de dados categóricos, gender, internet service, phone service, etc., terão que ser convertidos em colunas correspondentes contendo 0 ou 1. 

Dados categóricos como Sim / Não, bom / ruim / feio não podem ser usados ​​para operações matemáticas . Portanto, essas colunas contendo dados categóricos acabarão se tornando colunas com 0 ou 1.

Para dados numéricos contínuos, a escala de dados para cada coluna é diferente. 

Por exemplo, a coluna de idade tem valores que variam de 10 a 89, mas a coluna de salário tem números em torno de 10s ou 100s de milhares. 

Antes de podermos usar esses dados para análise, eles precisam ser normalizados para que os dados sejam centralizados em torno da média e sejam medidos em termos de seu desvio em relação à média. Isso é importante para a estabilidade numérica do modelo.

```python

# As colunas a seguir têm tipos de dados categóricos e precisam ser convertidas em variáveis dummy
dummy_variable_columns = ['State',
                          'Churn','Area','Marital','Gender','Techie','Contract','Port_modem','Tablet',
                         'InternetService','Phone', 'Multiple','OnlineSecurity','OnlineBackup',
                          'DeviceProtection','TechSupport', 'StreamingTV','StreamingMovies',
                          'PaperlessBilling','PaymentMethod'
                         ]


# As colunas a seguir possuem dados numéricos contínuos, que precisam ser normalizados antes que o modelo preditivo possa ser criado..
# Portanto, essas colunas serão transformadas para a forma normalizada.

to_transform = ['Population','Children','Age','Income','Outage_sec_perweek','Email',
                'Contacts','Yearly_equip_failure','Tenure',
                'MonthlyCharge','Bandwidth_GB_Year'
               ]

dummy_data_file_index = 0
```

```python
from sklearn.preprocessing import StandardScaler

def get_dummy_data_with_output(dummy_variable_columns, data):
    global dummy_data_file_index
    dummy_data = pd.get_dummies(data, prefix=dummy_variable_columns, columns=dummy_variable_columns, drop_first= True)
    dummy_data = column_name_remove_space(dummy_data)
    dummy_data.to_csv('dummy_var_data'+str(dummy_data_file_index)+'.csv', index=False)
    y = dummy_data['Churn_Yes']
    dummy_data.drop(columns=['Churn_Yes'], inplace=True)
    sc = StandardScaler()
    dummy_data.loc[:,to_transform] = sc.fit_transform(dummy_data.loc[:,to_transform])
    dummy_data_file_index += 1
    return y, dummy_data

y, dummy_data = get_dummy_data_with_output(dummy_variable_columns, data)
```

### Modelo Inicial

Como uma primeira etapa, para verificar o impacto, a importância e a significância de várias colunas de dados na análise de rotatividade, será criado um modelo inicial contendo todas as variáveis ​​no conjunto de dados.

Preparamos os dados na etapa anterior, de forma que inclui todas as variáveis ​​neste momento.

O modelo inicial fornecerá informações sobre quais variáveis ​​são significativas para a análise preditiva. Com base na significância das variáveis, devemos ser capazes de eliminar algumas variáveis ​​do conjunto de dados para chegar a um modelo reduzido. 

Em seguida, verificaremos a pontuação de precisão, a matriz de confusão e a AUC para os 2 modelos, para comparar seu desempenho.

Vamos começar importando as bibliotecas necessárias. Estamos usando bibliotecas como sklearn e modelos de estatísticas para criar modelos para verificar o quão bom o modelo preditivo funciona e quão precisamente somos capazes de prever os resultados ou a rotatividade do cliente. Como temos um grande número de linhas nos dados, vamos usar 70% dos dados para o conjunto de treinamento e 30% dos dados para o conjunto de teste.




```python
# importando as bibliotecas necessárias

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import statsmodels.api as sm
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
# separando os dados de treino e teste 7:3 
X_train, X_test, y_train, y_test = train_test_split(dummy_data,y,test_size=.3, random_state=0)
```

```python

# Função para criar um modelo de previsão com base na biblioteca de modelos estatísticos e
#imprimir detalhes como resumo do modelo, matriz de confusão e pontuação de precisão com base nos valores previstos usando o conjunto de teste

def get_sm_model(y_train,X_train, X_test):
    X_train = sm.add_constant(X_train)
    X_test = sm.add_constant(X_test)
    logit_model=sm.Logit(y_train,X_train)
    model=logit_model.fit()
    print(model.summary())
    y_hat = list(map(round, model.predict(X_test)))
    arr = np.c_[y_hat, y_test]
    print('\nValores previstos e reais:\n',arr)
    print('Confusion Matrix \n',confusion_matrix(y_test, y_hat))
    print('Accuracy Score: ',accuracy_score(y_test, y_hat))
    return model, arr

# Função para criar um modelo de previsão com base na biblioteca sklearn 
# e imprimir detalhes como Resumo do modelo, Matriz de confusão e Pontuação de precisão com base em valores previstos usando o conjunto de teste

def get_sk_model(y_train,X_train, X_test):
    model = LogisticRegression(max_iter=100000000)
    model.fit(X_train, y_train)
    arr = np.c_[X_train.columns.tolist(),model.coef_[0]]
    intercept = model.intercept_[0]
    print('\nCeficientes de modelo e resumo de interceptação para modelo sklearn:\n',arr, model.intercept_)
    y_pred = model.predict(X_test)
    print('\nValores previstos e reais do sklearn::\n',np.c_[y_pred, y_test])
    print('Confusion Matrix\n',confusion_matrix(y_test, y_pred))
    plt.matshow(confusion_matrix(y_test, y_pred))
    plt.title('Confusion matrix')
    plt.colorbar()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.grid(b=None)
    plt.show()
    print('Accuracy Scores:\n',accuracy_score(y_test, y_pred))
    print('Classification Report:\n',classification_report(y_test, y_pred))
    return model, arr, intercept


# Função para criar gráfico AUC e imprimir pontuação AUC. 
# A pontuação AUC mais próxima de 1 é considerada um bom modelo 
# Auc pontuação de 0,5 representa é um modelo que não é 
# melhor do que uma estimativa aleatória com 50% de chance de prever o resultado certo

from sklearn.metrics import roc_curve
def roc(model, X_test, y_test):
    probs = model.predict_proba(X_test)
    fpr, tpr, _ = roc_curve(y_test, probs[:,1])
    plt.plot(fpr, tpr, marker='.')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
    print('AUC: %.3f' % roc_auc_score(y_test, probs[:,1]))
```

```python
# Executar função de modelo 
get_sm_model(y_train,X_train, X_test);
model, arr, intercept = get_sk_model(y_train,X_train, X_test);
# Run AUC
roc(model, X_test, y_test);
```
