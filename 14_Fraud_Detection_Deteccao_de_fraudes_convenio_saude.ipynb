# **Introdução**

A fraude no seguro saúde é um dos maiores problemas que o Medicare enfrenta. 

Nesse tipo de fraude, informações falsas ou enganosas são fornecidas a uma seguradora de saúde ou ao governo na tentativa de fazer com que eles paguem benefícios não autorizados ao titular da apólice, outra parte ou entidade prestadora de serviços.

De acordo com o governo dos EUA, os gastos totais com o Medicare aumentaram exponencialmente devido a fraudes em reivindicações do Medicare. Fraude na área de saúde é um crime organizado que envolve colegas de prestadores, médicos e beneficiários agindo em conjunto para fazer reivindicações de fraude. 

Nesse crime, os provedores adotam formas nas quais um código de diagnóstico ambíguo é usado para adotar procedimentos e medicamentos mais caros. As seguradoras são as instituições mais vulneráveis ​​afetadas por essas práticas inadequadas. Por esse motivo, as seguradoras aumentaram seus prêmios de seguro e, como resultado, os cuidados de saúde estão se tornando uma questão cara a cada dia. A fraude e o abuso no setor de saúde assumem várias formas. Alguns dos tipos mais comuns de fraudes por parte dos provedores são:

* Faturamento de serviços que não foram fornecidos
* Envio duplicado de uma solicitação para o mesmo serviço
* Representar falsamente o serviço prestado
* Cobrar por um serviço mais complexo ou caro do que o realmente prestado
* Faturamento por um serviço coberto quando o serviço realmente prestado não estava coberto




## **Contexto:**

A **detecção de fraude** é um conjunto de atividades realizadas para evitar que dinheiro ou propriedade sejam obtidos por meio de falsos pretextos.
A fraude pode ser cometida de diferentes maneiras e em diferentes ambientes. Por exemplo, a fraude pode ser cometida nos setores de bancos, seguros, governo e saúde.


Um tipo comum de fraude bancária é a apropriação da conta do cliente. Isso ocorre quando alguém obtém acesso ilegal à conta bancária da vítima usando bots . Outros exemplos de fraude no setor bancário incluem o uso de aplicativos maliciosos, o uso de identidades falsas, lavagem de dinheiro , fraude de cartão de crédito e fraude móvel. Fraude de seguro inclui fraude de desvio de prêmio, que é o desvio de prêmios de seguro, ou rotatividade gratuita, que é negociação excessiva por um corretor da bolsa para maximizar as comissões. 

Outras formas de fraude de seguro incluem desvio de ativos, fraude de compensação de trabalhadores, fraude de acidente de carro, fraude de carro roubado ou danificado e fraude contra incêndio domiciliar. 

O motivo por trás de todas as fraudes de seguro são os lucros financeiros.
Fraude governamental consiste em cometer fraude contra agências federais, como o Departamento de Saúde e Serviços Humanos dos Estados Unidos, o Departamento de Transporte, o Departamento de Educação ou o Departamento de Energia. Os tipos de fraude governamental incluem cobrança de procedimentos desnecessários, cobrança excessiva de itens que custam menos, fornecimento de equipamento antigo ao cobrar por equipamento novo e relatórios de horas trabalhadas para um trabalhador que não existe.

Fraude de saúde inclui fraude de drogas e fraude médica, bem como fraude de seguro. A fraude na área de saúde é cometida quando alguém defrauda uma seguradora ou programa de saúde do governo.

A fraude pode ser cometida de diferentes maneiras e em muitos setores. A maioria dos métodos de detecção **combina uma variedade de conjuntos de dados** para formar uma visão geral conectada de dados de pagamento válidos e não válidos para tomar uma decisão. Essa decisão deve levar em consideração o endereço IP, geolocalização, identificação do dispositivo, dados “BIN”, latitude / longitude global, padrões históricos de transação e as informações reais da transação. Na prática, isso significa que os comerciantes e emissores implementam respostas com base analítica que usam dados internos e externos para aplicar um conjunto de regras de negócios ou algoritmos analíticos para detectar fraudes.

## Combinando todos os dados

O objetivo do projeto é prever se um provedor é fraudulento ou não com base nas reivindicações feitas por ele. Como os dados dos provedores, beneficiários e sinistros estão todos separados, torna-se necessário **combinar todos os dados** para ter o formato (xi, yi) para podermos usar os algoritmos de aprendizado de máquina supervisionado.

```python
def combine_data(provider, beneficiary, inpatient, outpatient):
    ''' Esta função combina dados de provedor, beneficiário, paciente interno e externo juntos em um único arquivo '''
    # Encontra as colunas comuns em dados de pacientes internados e dados ambulatoriais
    common_inp_outp = [col for col in inpatient if col in outpatient]
    # Junção dos dados de pacientes internos e externos usando colunas comuns
    train_in_out = outpatient.merge(inpatient, on=common_inp_outp, how='outer', indicator='is_inpatient')
    train_in_out = train_in_out.replace({'left_only':0, 'right_only':1})
    # Inner joining dos dados dos benéficiarios e combinação de paciente interno e externo
    train_benef_in_out = beneficiary.merge(train_in_out, on='BeneID')
    # combinando todas as combinações com os dados do trem contendo a variável de destino
    combined_data = provider.merge(train_benef_in_out, on='Provider')
    return combined_data
```

```python
# Combine todos os dados do conjunto de treinamento
train_data = combine_data(train, train_beneficiary, train_inpatient, train_outpatient)
print("Shape: ", train_data.shape)
train_data.head()
```

## Data pre-processing

Como os dados passarão para um modelo de aprendizado de máquina, é **obrigatório converter cada coluna para o formato correto**, por exemplo: o tipo de dados da variável de destino PotentialFraud é objeto, pois contém valores Sim e Não, então eles devem ser substituídos por 1 e 0 respectivamente. Da mesma forma, o tipo de dados das colunas de data precisa ser alterado de objeto para data e hora.

```python
# 1. Substituindo Sim por 1 e Não por 0 na coluna PotentialFraud
train_data['PotentialFraud'] = train_data['PotentialFraud'].replace({'Yes':1, 'No':0})
train_data.PotentialFraud.value_counts()
```

```python
# 2. Alterar DOB e DOD para o tipo de data
train_data['DOB'] = pd.to_datetime(train_data['DOB'])
train_data['DOD'] = pd.to_datetime(train_data['DOD'])
train_data.head()
```

```python
# Contagem de valor da coluna de gênero
train_data.groupby('Gender')['BeneID'].nunique().plot.bar()
plt.title("Reivindicações feitas para cada gênero")
plt.show()
```

É muito provável que mais o número maior de reivindicações seja feitas para os homens, à medida que mais homens adoecem a cada ano.

```python
# 3. Substituindo 1 por Feminino e 2 por Masculino
train_data['Gender'] = train_data['Gender'].replace({1:'Female', 2:'Male'})
train_data.head()
```

```python
# 4. Substituindo '0' por 0 e 'Y' por 1 na coluna RenalDiseaseIndicator
train_data.RenalDiseaseIndicator.replace({'0':0, 'Y':1}, inplace=True)
train_data.head()
```

```python
# 5. Substituindo 2 por 0 para todas as condições crônicas, por exemplo: ChronicCond_Alzheimer
train_data.replace({'ChronicCond_Alzheimer': 2, 'ChronicCond_Heartfailure': 2, 'ChronicCond_KidneyDisease': 2,
                           'ChronicCond_Cancer': 2, 'ChronicCond_ObstrPulmonary': 2, 'ChronicCond_Depression': 2, 
                           'ChronicCond_Diabetes': 2, 'ChronicCond_IschemicHeart': 2, 'ChronicCond_Osteoporasis': 2, 
                           'ChronicCond_rheumatoidarthritis': 2, 'ChronicCond_stroke': 2 }, 0, inplace=True)
train_data.iloc[:, 12:22].head(10)
```

```python
# 6. Alterando o tipo de dados de ClaimStartDt e ClainEndDt para datetime
train_data['ClaimStartDt'] = pd.to_datetime(train_data['ClaimStartDt'])
train_data['ClaimEndDt'] = pd.to_datetime(train_data['ClaimEndDt'])
```

```python
# 7. Alterar o tipo de dados de AdmissionDt e DischargeDt para datetime
train_data['AdmissionDt'] = pd.to_datetime(train_data['AdmissionDt'])
train_data['DischargeDt'] = pd.to_datetime(train_data['DischargeDt'])
```

## Adicionando novos recursos

As features existentes no conjunto de dados podem não ser suficientes para prever se um provedor de saúde é uma fraude ou não. Por iss, aqui estão outras features que podem ajudar na identificação de fornecedores fraudulentos:

* is_dead: Esta é uma coluna de sinalizador para indicar se o beneficiário estava morto quando a reclamação foi feita. Podemos simplesmente substituir os valores não nulos por 1 e os valores nulos por 0 na coluna DOD, uma vez que a ausência de valor na coluna DOD deve significar que a pessoa estava viva quando a reclamação foi feita.

* Days_Admitted: indica o No. de dias em que o beneficiário esteve internado no hospital e é calculado subtraindo-se a data de admissão da data de alta.

* Diseases_Count: indica o número total de doenças crônicas que um beneficiário que fez a solicitação tinha.

* Total_Physicians: indica o número total de médicos que trataram ou atenderam a um beneficiário.

* Total_ClaimCodes: indica o número total de códigos de diagnóstico de sinistro usados ​​na fatura médica feita pelo provedor.

* Total_ProcedureCodes: indica o número total de procedimentos médicos usados ​​em um beneficiário que dá uma estimativa de custo incorrido pelo beneficiário.

* Claim_Period: indica o período de reclamação de um beneficiário, que é a diferença entre ClaimStartDt e ClaimEndDt.

* ClaimPeriod_greater_DaysAdmitted : indica se o período de reclamação é maior que No. de dias um paciente foi admitido no hospital.

* is_claimcode: Esta é uma coluna de sinalizador que indica a presença de ClmAdmitDiagnosisCode em uma reivindicação com 1 significando Sim e 0 significando não.

* is_groupcode: Esta é uma coluna de sinalizador que indica a presença de DiagnosisGroupCode em uma reivindicação com 1 significando Sim e 0 significando Não.

## Modelagem

Vários modelos de máquina foram testados com os dados finais e abaixo está uma análise do código com os resultados das métricas de avaliação escolhidas.

### Métricas de Avaliação

Uma vez que o custo da classificação incorreta é alto e não podemos ter:

* Altos Falsos Negativos: Provedor fraudulento classificado incorretamente como provedor genuíno, ou seja, baixo recall

* Altos Falsos Positivos: Fornecedor genuíno mal classificado como fornecedor fraudulento, ou seja, baixa precisão

* No entanto, nosso objetivo principal é ter menos No.. de falsos negativos, mesmo que isso signifique ter falsos positivos ligeiramente elevados, pois a não detecção de um provedor fraudulento pode causar enormes perdas para as seguradoras, o que, a longo prazo, tornaria os cuidados de saúde mais caros.

Podemos usar Recall, pontuação F1 e pontuação AUC como métricas de avaliação.

## Train-Test split (80:20)
