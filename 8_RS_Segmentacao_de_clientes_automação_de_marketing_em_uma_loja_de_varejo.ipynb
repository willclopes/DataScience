## Segmentação do Cliente

### RFM

Um das formas de conhecer o cliente é realizar o RFM, que é uma das formas mais simples e rápidas de conhecer o seu cliente. 

RFM é um método usado para analisar o valor do cliente. É comumente usado em marketing de banco de dados e marketing direto e tem recebido atenção especial nos setores de varejo e serviços profissionais.

RFM significa as três dimensões:

* Tempo para retorno - há quanto tempo o cliente comprou?
* Frequência - com que frequência eles compram?
* Valor monetário - quanto eles gastam?

Existem também outras variantes desta técnica:

* **RFD - Recência, frequência, duração** é uma versão modificada da análise RFM que pode ser usada para analisar o comportamento do consumidor de produtos de negócios orientados para visualização / leitura / navegação.

* **RFE - Recência, frequência, envolvimento** é uma versão mais ampla da análise RFD, onde o envolvimento pode ser definido para incluir a duração da visita, páginas por visita ou outras métricas.

* **RFM-I - Recência, Frequência, Valor Monetário - Interações** é uma versão da estrutura RFM modificada para contabilizar a recência e a frequência das interações de marketing com o cliente (por exemplo, para controlar possíveis efeitos dissuasores de engajamentos de publicidade muito frequentes).

* **RFMTC - Recência, Frequência, Valor Monetário, Tempo, Taxa de Churn** um modelo RFM aumentado proposto por I-Cheng et al. (2009). O modelo utiliza a sequência de Bernoulli na teoria da probabilidade e cria fórmulas que calculam a probabilidade de um cliente comprar na próxima campanha promocional ou de marketing. O modelo foi implementado por Alexandros Ioannidis para conjuntos de dados como a Transfusão de Sangue e conjuntos de dados CDNOW.

[Referência](https://en.wikipedia.org/wiki/RFM_(market_research))

A primeira coisa que precisamos fazer é calcular três valores - Recência, Frequência e o valor gasto para cada cliente no conjunto de dados por algum período de tempo. Vamos fazer esse cálculo com base no semestre para os 3 últimos períodos.

A próxima etapa é criar três grupos diferentes com base no Recência, Frequência e Valor Monetário dos clientes, o que nos ajudaria a nomear nossos segmentos de clientes. 

A pontuação RFM é uma soma simples de Recência, Frequência e Valores Monetários e essa soma é o resultado de valores inteiros como 10, 9, 8 e etc.

Esta pontuação indicará o valor da pontuação RFM que nos permitirá tomar decisões em um produto comercial ou em nossos clientes. Esta é uma métrica muito importante devido aos futuros processos de tomada de decisão em relação aos usuários ou clientes.

No final, temos a análise RFM para três períodos de tempo.


### Calculando valores RFM

A primeira coisa que vamos fazer é encontrar o valor de "recência", o número de dias desde o último pedido de um cliente e essas informações estarão na coluna **InvoiceData**. 

Nessa coluna, podemos saber quando foi a primeira compra e quando foi a última compra de um cliente.

### Valores RFM

RFM (Recência, Frequência, Monetária) é uma Técnica muito Simples que podemos aplicá-la muito facilmente e obter uma análise super útil para a nossa Segmentação de Clientes. O tempo para retorno corresponde a dias desde que os clientes fizeram a última compra e, por definição, quanto menor, melhor. A frequência é o número de transações nos últimos 12 meses. O valor monetário é a quantia total de dinheiro que os clientes gastaram nos últimos 12 meses.



Como escolhemos aleatoriamente o subconjunto dos dados, precisamos saber o período de tempo de nossos dados. Como o que você pode ver abaixo, o último dia de nosso conjunto de dados é 9 de dezembro de 2011. Portanto, defina 10 de dezembro como nossa data final e conte retroativamente o número de dias a partir da última compra de cada cliente. Esse será o valor de recência.

```python
print('Min : {}, Max : {}'.format(min(online.InvoiceDay), max(online.InvoiceDay)))
```

```python
#fixar a ultima data
pin_date = max(online.InvoiceDay) + dt.timedelta(1)
```

```python
pin_date
```

Antes de obter o recência, vamos preparar mais um valor com antecedência, a quantia total de dinheiro que cada cliente gastou. Isso serve para contar o valor monetário. Como podemos conseguir isso? Fácil! Multiplicando o preço do produto e a quantidade do pedido em cada linha.

```python
# Cria uma coluna de gasto total 
online['TotalSum'] = online.Quantity * online.UnitPrice 
online.head ()
```

Agora estamos prontos para obter os três valores RFM de uma vez. 

Vamos agrupar os dados de cada cliente e agregá-los para cada recência, frequência e valor monetário.

```python
# calculando os valores de RFM
rfm = online.groupby('CustomerID').agg({
    'InvoiceDate' : lambda x: (pin_date - x.max()).days,
    'InvoiceNo' : 'count', 
    'TotalSum' : 'sum'})
```

```python

# renomeando as colunas
rfm.rename(columns = {'InvoiceDate' : 'Recency', 
                      'InvoiceNo' : 'Frequency', 
                      'TotalSum' : 'Monetary'}, inplace = True)

rfm.head()
```

### Quartis RFM

Agora vamos agrupar os clientes com base nos valores RFM. Como esses valores são contínuos, também podemos usar os valores dos quantis e dividi-los em 4 grupos.


```python
# criando rótulos e atribuindo aos grupos de percentis 
r_labels = range(4, 0, -1)
r_groups = pd.qcut(rfm.Recency, q = 4, labels = r_labels)

f_labels = range(1, 5)
f_groups = pd.qcut(rfm.Frequency, q = 4, labels = f_labels)

m_labels = range(1, 5)
m_groups = pd.qcut(rfm.Monetary, q = 4, labels = m_labels)

m_groups.head()
```

Uma atenção extra para o r_labels. 

Os "labels" estão em ordem decrescente, isso Por que a recência significa quanto tempo passou desde o último pedido de um cliente. Portanto, quanto menor o valor, mais engajado o cliente está com a marca. Agora vamos fazer uma nova coluna para indicar os rótulos dos grupos.

```python
# crie uma nova coluna para rótulos de grupo 
rfm['R'] = r_groups.values
rfm['F'] = f_groups.values
rfm['M'] = m_groups.values

# some as três colunas 
rfm['RFM_Segment'] = rfm.apply(lambda x: str(x['R']) + str(x['F']) + str(x['M']), axis = 1)
rfm['RFM_Score'] = rfm[['R', 'F', 'M']].sum(axis = 1)
rfm.head()
```

Concatenamos todos os três rótulos em uma célula como RFM_Segment. 

Desta forma, podemos verificar facilmente a que nível ou segmento um cliente pertence. 

O RFM_Score é a soma total dos três valores. 

Não precisa ser necessariamente a soma, então o valor médio também é possível. 


Além disso, podemos obter padrões adicionais com a média ou valores de contagem de recência, frequência e monetário agrupados por essa pontuação, como abaixo.


```python
#calculando os valores médios para cada RFM  
rfm_agg = rfm.groupby('RFM_Score').agg({
    'Recency' : 'mean',
    'Frequency' : 'mean',
    'Monetary' : ['mean', 'count']
})

rfm_agg.round(1).head()
```

RFM_Scoreserá a pontuação total do envolvimento ou lealdade de um cliente. 

Resumindo os três valores juntos, podemos finalmente categorizar os clientes em 'Ouro', 'Prata', 'Bronze' e 'Verde'.

```python
# atribuindo rótulos da pontuação total 
score_labels = ['Green', 'Bronze', 'Silver', 'Gold']
score_groups = pd.qcut(rfm.RFM_Score, q = 4, labels = score_labels)
rfm['RFM_Level'] = score_groups.values

rfm.head()
```

Excelente! Terminamos com uma análise de corte com valores RFM. 

Identificamos quem é a nossa galinha dos ovos de ouro e onde devemos tomar cuidado redobrado.

## Agrupamento K-Means

O agrupamento K-Means é um tipo de algoritmo de aprendizagem não supervisionado, que cria grupos com base na distância entre os pontos. 
Como? Existem dois conceitos de distância no agrupamento K-Means. Dentro das somas dos quadrados dos agrupamentos (WSS) e entre as somas dos quadrados dos agrupamentos (BSS).


image.png

WSS significa a soma das distâncias entre os pontos e os centróides correspondentes para cada cluster e BSS significa a soma das distâncias entre os centróides e a média total da amostra multiplicada pelo número de pontos dentro de cada cluster. Portanto, você pode considerar WSS como a medida de compactação e BSS como a medida de separação. Para que o cluster seja bem-sucedido, precisamos obter o WSS inferior e o BSS superior.

Ao iterar e mover os centróides do cluster, o algoritmo K-Means tenta obter os pontos otimizados do centróide, que minimizam o valor do WSS e maximizam o valor do BSS.

Como o clustering K-means usa a distância como fator de similaridade, precisamos normalizar os dados. Suponha que temos duas escalas diferentes de recursos, digamos, altura e peso. A altura é superior a 150 cm e o peso é inferior a 100 kg em média. Portanto, se plotarmos esses dados, a distância entre os pontos será altamente dominada pela altura, resultando em uma análise tendenciosa.

```python
# plotando os valores atuais de RFM
plt.subplot(3, 1, 1); sns.distplot(rfm.Recency, label = 'Recency')
plt.subplot(3, 1, 2); sns.distplot(rfm.Frequency, label = 'Frequency')
plt.subplot(3, 1, 3); sns.distplot(rfm.Monetary, label = 'Monetary')

plt.tight_layout()
plt.show();
```

Como pode-se ver acima, os valores estão distorcidos e precisam ser normalizados.

Portanto, quando se trata de clustering K-means, patronizar/normalizar os dados é uma etapa crítica para o pré-processamento. Se verificarmos a distribuição dos valores RFM, você pode notar que eles estão distorcidos para a direita. Não é um bom estado para usar sem padronização. Vamos transformar os valores RFM em log em escala primeiro e depois normalizá-los. 

Precisamos defini-los como 1 antes da transformação do log e do escalonamento.

```python
# define a função para os valores abaixo de 0 
def neg_to_zero(x):
    if x <= 0:
        return 1
    else:
        return x

# aplique a função à coluna Recência e MonetaryValue 
rfm['Recency'] = [neg_to_zero(x) for x in rfm.Recency]
rfm['Monetary'] = [neg_to_zero(x) for x in rfm.Monetary]
rfm.head()
```

```python
rfm_log = rfm[['Recency', 'Frequency', 'Monetary']].apply(np.log, axis = 1).round(3)
rfm_log.head()
```

```python
rfm_log.describe()
```

A próxima etapa de pré-processamento é a padronização, mas é mais simples do que a etapa anterior. Usando StandardScaler() , podemos obter os valores padronizados como abaixo.

```python
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)

# transformando em um dataframe
rfm_scaled = pd.DataFrame(rfm_scaled, index = rfm.index, columns = rfm_log.columns)
rfm_scaled.head()
```

```python
# plotando a nova distribuíção doso valores de RFM 
plt.subplot(3, 1, 1); sns.distplot(rfm_scaled.Recency, label = 'Recency')
plt.subplot(3, 1, 2); sns.distplot(rfm_scaled.Frequency, label = 'Frequency')
plt.subplot(3, 1, 3); sns.distplot(rfm_scaled.Monetary, label = 'Monetary')

plt.tight_layout()
plt.show()
```

Agora, terminamos o pré-processamento.

Qual é a próxima? A próxima etapa será selecionar o número certo de clusters (K). Temos que escolher quantos grupos vamos formar. Se houver conhecimento prévio, podemos simplesmente fornecer o número à frente para o algoritmo. Mas, na maioria dos casos, no aprendizado não supervisionado, não existe. Portanto, precisamos escolher o número otimizado, e o método Elbow é uma das soluções onde podemos obter as dicas.

Com o método Elbow, podemos obter o número ideal de clusters.

```python
wcss = {}

for i in range(1, 11):
    kmeans = KMeans(n_clusters= i, init= 'k-means++', max_iter= 300)
    kmeans.fit(rfm_scaled)
    wcss[i] = kmeans.inertia_
```

Usando o loop for, constituímos os modelos para cada número de clusters de 1 a 10. E, em seguida, foi coletado os valores de WSS para cada modelo. Veja o gráfico abaixo. Conforme o número de clusters aumenta, o valor do WSS diminui. Não há surpresa porque quanto mais clusters fizermos, o tamanho de cada cluster diminuirá, de modo que a soma das distâncias dentro de cada cluster diminuirá. Então, qual é o número ideal?

```python
sns.pointplot(x = list(wcss.keys()), y = list(wcss.values()))
plt.xlabel('K Numbers')
plt.ylabel('WCSS')
plt.show()
```

A resposta está no 'cotovelo' desta linha. Em algum lugar o WSS diminui drasticamente, mas não muito K. Escolhemos aqui o três. O que você me diz? Não parece realmente o "cotovelo" da linha?

Agora que escolhemos o número de clusters, podemos construir um modelo e fazer clusters reais. Também podemos verificar a distância entre cada ponto e os centróides ou os rótulos dos clusters. Vamos fazer uma nova coluna e atribuir os rótulos a cada cliente.

```python
# treinando o modelo
clus = KMeans(n_clusters= 3, init= 'k-means++', max_iter= 300)
clus.fit(rfm_scaled)
```

```python
 # criando uma nova coluna com o número dos os clusters definido pelo modelo
rfm['K_Cluster'] = clus.labels_
rfm.head()
```

Agora fizemos dois tipos de segmentação, grupos de quantis RFM e grupos K-Means. Vamos fazer a visualização e comparar os dois métodos.


## Comparando os métodos

Vamos fazer dois tipos de gráfico, um gráfico de linha e um mapa de calor. Podemos facilmente comparar as diferenças dos valores de RFM com esses dois gráficos. Em primeiro lugar, criaremos colunas que serão o valores de RFM e K-means. Em seguida, vamos remodelar os dados, fundindo os valores de RFM em uma coluna.

```python
#atribuindo novas colunas
rfm_scaled['K_Cluster'] = clus.labels_
rfm_scaled['RFM_Level'] = rfm.RFM_Level
rfm_scaled.reset_index(inplace = True)
rfm_scaled.head()
```

```python
# melt the dataframe
rfm_melted = pd.melt(frame= rfm_scaled, id_vars= ['CustomerID', 'RFM_Level', 'K_Cluster'], 
                     var_name = 'Metrics', value_name = 'Value')
rfm_melted.head()
```

```python
sns.lineplot(x = 'Metrics', y = 'Value', hue = 'RFM_Level', data = rfm_melted)
plt.title('Snake Plot of RFM')
plt.legend(loc = 'upper right')
```

```python
# a snake plot with K-Means
sns.lineplot(x = 'Metrics', y = 'Value', hue = 'K_Cluster', data = rfm_melted)
plt.title('Snake Plot of K_cluster')
plt.legend(loc = 'upper right')
```

Esse tipo de plotagem é chamado de 'Snake plot', especialmente na análise de marketing. Parece que os grupos Ouro e Verde no gráfico esquerdo são semelhantes com 1 e 2 clusters no gráfico direito. E os grupos Bronze e Silver parecem fundidos no grupo 0 .

### Mapa de Calor

Vamos tentar novamente com um mapa de calor. Os mapas de calor são uma representação gráfica dos dados em que valores maiores foram coloridos em escalas mais escuras e valores menores em escalas mais claras. Podemos comparar a variação entre os grupos de forma bastante intuitiva por cores.

```python
# calcula o valor médio no total 
total_avg = rfm.iloc[:, 0:3].mean()
total_avg

# calcula uma coluna proporcional com média total 
cluster_avg = rfm.groupby('RFM_Level').mean().iloc[:, 0:3]
prop_rfm = cluster_avg/total_avg - 1

# mapara de calor com RFM
sns.heatmap(prop_rfm, cmap= 'Oranges', fmt= '.2f', annot = True)
plt.title('Heatmap of RFM quantile')
plt.plot()
```

Em seguida, repita o mesmo código para clusters K como fizemos antes.


```python
#calcula a coluna proporcional com a média total 
cluster_avg_K = rfm.groupby('K_Cluster').mean().iloc[:, 0:3]
prop_rfm_K = cluster_avg_K/total_avg - 1

# mapa de calor com K-means
sns.heatmap(prop_rfm_K, cmap= 'Blues', fmt= '.2f', annot = True)
plt.title('Heatmap of K-Means')
plt.plot()
```
